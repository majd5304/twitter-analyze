```python
import pandas as pd
data = pd.read_csv('/content/Dataset_Amini.csv')
data["ID"] = ""
```

```python
heads=list(data['Head'].apply(eval))
heads_data=pd.DataFrame(heads)
heads_id=list(set(heads_data['id']))
```

```python
from tqdm import tqdm 
```

```python
def joindic(lst):
  diclist=[]
  for x in tqdm(lst):
    diclist.extend(x)
  return diclist

exterlink=list(data['ExternalLinks'].apply(eval))
external_links=joindic(exterlink)

extenal_data=pd.DataFrame(external_links)
external_id=list(set(extenal_data['id']))
```

```python
external_id
```

```python
def intersection(lst1,lst2):
  lst3=[value for value in lst1 if value in lst2]
  return lst3


intersection_ids=intersection(heads_id,external_id)
```

```python
intersection_ids
```

```python
import re
def maining_hashtag(text):
  return re.findall(r'#(\w+)', text)

data["Text_hashtag"] = data["Text"].apply(maining_hashtag)
```

```python
data["Text_hashtag"]
```

```python
heads=list(data['Head'].apply(eval))
heads_data=pd.DataFrame(heads)
data['heads_id']=list((heads_data['id']))
```

```python
data_new=data[['heads_id','Text_hashtag']]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# create an empty graph
G = nx.Graph()

# add nodes for the IDs and hashtags
ids = ['@johndoe', '@janedoe', '@mark', '@kate']
hashtags = ['#OpIran', '#اعتصابات_سراسری', '#مهسا_امینی']

G.add_nodes_from(ids, bipartite=0)
G.add_nodes_from(hashtags, bipartite=1)

# add edges between IDs and hashtags
edges = [('@johndoe', '#OpIran'), ('@johndoe', '#مهسا_امینی'),
         ('@janedoe', '#اعتصابات_سراسری'), ('@mark', '#OpIran'),
         ('@mark', '#مهسا_امینی'), ('@kate', '#OpIran')]

G.add_edges_from(edges)

# draw the graph
pos = nx.bipartite_layout(G, ids)
nx.draw_networkx_nodes(G, pos, nodelist=ids, node_color='r')
nx.draw_networkx_nodes(G, pos, nodelist=hashtags, node_color='b')
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_labels(G, pos)

plt.axis('off')
plt.show()
```
